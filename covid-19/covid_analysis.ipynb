## 확진자가 가장 많은 나라 순위
1. US
2. Brazil
3. India
4. Russia
5. South Africa

```python
df['D/C'] = (df['Deaths']/df['Confirmed'])
df.sort_values('D/C',ascending=False).head(5)
```

## 사망률이 가장 높은 나라
1. Yemen
2. United Kingdom
3. Belgium
4. Italy
5. France

```python
df['R/C'] = (df['Recovered']/df['Confirmed'])
df.sort_values('R/C',ascending=False).head(5)
```

 회복률이 가장 높은 나라를 조사하려 하였으나 Holy see 같은 나라는 표본이 너무 작아서 부적절하다고 판단

```python
df[df['Confirmed']>1000].sort_values('R/C',ascending=False).head(5)
```

확진자가 1000명이 넘는 국가들을 대상으로 재조사

## 회복률이 가장 높은 나라 순위
1. Djibouti
2. Iceland
3. New Zealand
4. Qatar
5. Malaysia

```python
df = df[df['Confirmed']>1000].sort_values('R/C',ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family']='Malgun Gothic'  # 한글 폰트
plt.rcParams['axes.unicode_minus']=False  # 마이너스(-) 오류 해결

plt.figure(figsize=(10,4))
plt.grid()
plt.title('회복률이 높은 나라 [막대그래프]',fontsize=17,color='g')
plt.bar(df['Country/Region'][:20],df['R/C'][:20],color='g')
```
