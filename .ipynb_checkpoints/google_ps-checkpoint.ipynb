```python
# 원본 자료는 자료가 너무 많아서 로딩하는 시간이 오래걸리기 때문에 일부만 가져왔습니다.

import pandas as pd
import glob
import matplotlib.pyplot as plt

path = './app_list/*xlsx'
file_list = glob.glob(path)
data_list=[]
for file in file_list:
    df = pd.read_excel(file)
    data_list.append(df)
data_list
```

```python
df = pd.concat(data_list)
df
```

```python
df.info()
```

### 앱 종류 중에 가장 많은 비중을 차지하는 앱 찾기

```python
df.groupby('Category').size().sort_values(ascending=False)
```

```python
game = pd.read_excel('./app_list/game.xlsx')
game.iloc[65:75]
```

```python
game['Rating'].fillna(0) #평점에 Nan값 0으로 채우기
```

## 더 나아가 보기
플레이스토어에서 미지원하는게 종류별 게임 순위 찾는 것이다. 
따라서 가장 리뷰가 많은 게임의 종류를 살펴보고 그 중에 무료 전략게임 순위를 알고 싶을 때 아래와 같은 코드를 사용할 수 있다.

#### 사람들이 관심있어 하는 게임 장르는?

```python
game.groupby(['Genres']).sum()['Reviews'].sort_values(ascending=False)
```

### 관심있어 하는 게임 장르 순위

1등: Action

2등: Casual

3등: Sports

### 상위 5개 원형그래프로 나타내기

```python
sizes = [214159660,189685412,122047448,118299680,81773507]
labels = 'Action','Casual','Sports','Arcade','Strategy'

fig1,ax1 = plt.subplots()
ax1.pie(sizes,labels=labels,autopct='%1.1f%%')
ax1.axis('equal')
plt.show()
```

### Strategy(전략)게임 중에 무료게임 순위 구하기

```python
gen_stra = game['Genres'] == 'Strategy'
fee_free = game['Price'] == 0.0
game[gen_stra&fee_free].sort_values('Reviews',ascending=False).head(5) #무료 전략 게임만 모으기
```
